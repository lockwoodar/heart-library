````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# load csv data from modified condastats output
# condastats overall heart-library --monthly
# data is broken and poorly formatted, requires some hand editing but is easy to shape with awk

df_heart = pd.read_csv("heart_data.csv", delim_whitespace=True)
print(df_heart)
print("\n\n")

# change to art_data_full.csv for more statistics
df_art = pd.read_csv("art_data.csv", delim_whitespace=True)
print(df_art)
````

````python
# convert cf data to datetime
df_heart["time"] = pd.to_datetime(df_heart["time"], format="%Y-%m")
df_art["time"] = pd.to_datetime(df_art["time"], format="%Y-%m")
print(df_heart)
print("\n\n")
print(df_art)
````

````python
# Create HEART data plot
plt.figure(figsize=(10,6))
plt.plot(df_heart["time"], df_heart["count"], marker="o", linestyle="-")
plt.title("heart-library Monthly Downloads via Conda-Forge")
plt.xlabel("Month")
plt.ylabel("Number of Downloads")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create ART-Core data plot
plt.figure(figsize=(10,6))
plt.plot(df_art["time"], df_art["count"], marker="o", linestyle="-")
plt.title("Adversarial-Robustness-Toolbox Monthly Downloads via Conda-Forge")
plt.xlabel("Month")
plt.ylabel("Number of Downloads")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
